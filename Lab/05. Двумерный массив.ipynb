```python
def print_matrix(A):
    for i in A:
        print(*[f"{x:>5}" for x in i])
    
def simple(number):
    n = number
    counter = 0
    if n == 1:
        return True
    else:
        for i in range(2, n//2):
            if n % i == 0:
                counter += 1
        return True if counter == 0 else False
```

2.	Дана матрица размера M × N. Удалить ее первую строку, содержащую только четные элементы.
3.	Дана матрица размера M × N. Вставить в него строку из нулей после первой из строк, содержащей хотя бы одно простое число.


```python
import copy
N = int(input("Введите количество столбцов: "))
M = int(input("Введите количество строк: "))
A = []

for i in range(M):
   A.append([random.randint(-100,100) for _ in range(N)])

print_matrix(A)

min_matr = A[0][0]
max_matr = A[0][0]
for i in range(M):
    if min_matr > min(A[i]): 
        min_matr = min(A[i])
        min_str = i
    if max_matr < max(A[i]): 
        max_matr = max(A[i])
        max_str = i
        
print('\nМинимальное значение: {}\tМаксимальное значение: {}'.format(min_matr, max_matr))
print('Минимальное значение в строке: {}\tМаксимальное значение в строке: {}\n'.format(min_str, max_str))
if min_str != max_str:
    buf_str = copy.deepcopy(A[min_str])
    A[min_str] = copy.deepcopy(A[max_str])
    A[max_str] = buf_str   

print_matrix(A)
```

```python
#Дана матрица размера M × N. Вставить в него строку из нулей после первой из строк, содержащей хотя бы одно простое число.
N = int(input("Введите количество столбцов: "))
M = int(input("Введите количество строк: "))
A = []

for i in range(M):
   A.append([random.randint(-100,100) for _ in range(N)])

# Для теста:
N = 4
M = 5
A = [[4,7,8,9],
     [1,3,5,7],
     [11,13,17,19],
     [4,6,10,12],
     [9,7,9,10]]

print_matrix(A)

for i in range(M):
    flag = 0
    first_str_simple = -1
    for j in range(N):
        if simple(A[i][j]):
            flag = 1
            first_str_simple = i
        else:
            flag = 0
            first_str_simple = -1
            break
    if flag == 1:
        break

if first_str_simple >= 0:
    print("\n", A[first_str_simple], "\n")
    A.insert(first_str_simple+1, [0]*N)
    print_matrix(A)
else: print("Такой строки нет")
```

```python
#Дана целочисленная матрица A(N, N), заполненная целыми числами. Отсортировать по убыванию элементы матрицы, от первого элемента до первого максимального элемента матрицы.
N = int(input("Введите размер квадратной матрицы: "))
A = []
for i in range(N):
   A.append([random.randint(-100,100) for _ in range(N)])

print_matrix(A)
print()

max_matr = A[0][0]
for i in range(N):
    for j in range(N):
        if A[i][j] > max_matr:
            max_matr = A[i][j]
            i_max = i
            j_max = j

for i in range(i):
    flag = 0
    for j in range(j):
        if A[i][j] == max_matr:
            i_max = i
            j_max = j
            flag = 1
            break
    if flag == 1: break

buf_matr = []
for i in range(N):
    if i < i_max:
        for j in range(N):
            buf_matr.append(A[i][j])
    elif i == i_max:
        for j in range(j_max):
            buf_matr.append(A[i][j])
    else:
        for j in range(j_max):
            buf_matr.append(A[i][j])
print("Строка с макисмумом = {}\nСтолбец с макисмумом = {}\nЗначение максимума = {}\nЭлементы матрицы от первого до максимума = {}".format(i_max+1, j_max+1, max_matr, buf_matr))
buf_matr.sort(reverse = True)
print("Сортированный массив =", buf_matr, '\n')
k = 0
for i in range(N):
    if i < i_max:
        for j in range(N):
            A[i][j] = buf_matr[k]
            k += 1
    elif i == i_max:
        for j in range(j_max):
            A[i][j] = buf_matr[k]
            k += 1
    else:
        for j in range(j_max):
            A[i][j] = buf_matr[k]
            k += 1

print_matrix(A)
```
